**Operate overloading**

```python
class point:
  def __init__(self,x,y):
    self.x = x
    self.y = y

  def __add__(self,other):
    x = self.x + other.x
    y = self.y + other.y
    return point(x,y)

p1 = point(2,3)
p2 = point(4,5)

print(p1+p2)
```

```python
class point:
  def __init__(self,x,y):
    self.x = x
    self.y = y

  def __add__(self,other):
    x = self.x + other.x
    y = self.y + other.y
    return point(x,y)

  def __str__(self):
    return f"{0},{1}".format(self.x,self.y)

p1 = point(2,3)
p2 = point(4,5)

print(p1+p2)
```

**Encapsulation:**
Encapsulation binds together data and functions that amnipulate the data, keeping it safe from interference and misuse.Encapsulation is one of the fundamental concepts in object-oriented programming(OOP).it describes the idea of wrapping data and the methods that work on data within one unit. This puts restrictions on accessing variables and methods directly and can prevent the accidental modification of data. To prevent accidental change, an object variable can only be changed by an objects method.Those types of variables are known as private variables. **self. PUBLIC

**self._ PROTECTED

**self.__ PRIVATE

self.name= public

self._name= protected/private

self.__name= strongly private

```python
class Encapsulation():
  def __init__(self,a,b,c):
    self.a=a
    self._b=b
    self.__c=c

  def mymethod(self):
    print(self.__c)
x= Encapsulation(11,15,17)
x.mymethod()

```
